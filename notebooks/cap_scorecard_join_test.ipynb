```python
# Select columns from scorecard to keep. EARN_NE_MDN_3YR provides median earnings 3 years after completion for non-enrolled students.
scorecard_fos_clean = scorecard_fos[[
    "UNITID"
    ,"INSTNM"
    ,"CIPCODE"
    ,"CIPDESC"
    ,"CREDLEV"
    ,"CREDDESC"
    ,"EARN_MDN_1YR"
    ,"EARN_NE_MDN_3YR"
    ,"EARN_MDN_5YR"
]].copy()
```

```python
desc_map = {
    "Undergraduate Certificate or Diploma" : "Certificate"
    ,"Associate's Degree" : "Associate"
    ,"Bachelor's Degree" : "Bachelor"
}
scorecard_tn["cred_desc"] = scorecard_tn["cred_desc"].map(desc_map)
```

```python
# Convert dtypes to prepare for PBI
scorecard_tn["unitid"] = scorecard_tn["unitid"].astype(str)
scorecard_tn["cip4"] = scorecard_tn["cip4"].astype(str)
scorecard_tn["credlev"] = scorecard_tn["credlev"].astype(str)
scorecard_tn["control"] = scorecard_tn["control"].astype(str)
```

```python
# Handle NaN and PS (Privacy Suppressed) values
cols_to_clean = ["median_earnings_1yr", "median_earnings_3yr", "median_earnings_5yr"]

for col in cols_to_clean:
    scorecard_tn[col] = (
        scorecard_tn[col]
            .replace("PS", np.nan)
            .replace({",": "", "\$": ""}, regex=True)
            .pipe(pd.to_numeric, errors="coerce")
            .astype("float64")
    )
```

```python
# Drop columns before exporting to PBI.
scorecard_tn_export = scorecard_tn.drop(columns=["cipcode", "avg_annual_tuition"])
scorecard_tn_export.head(5)
```

```python
#####
def normalize_unitid(series):
    return series.astype(str).str.replace('.0','', regex=False).str.strip()
```

```python
#####
scorecard_tn_export['unitid_n'] = (
    scorecard_tn_export['unitid']
    .astype(str)
    .str.replace('.0', '', regex=False)
    .str.strip()
)
# scorecard_tn_export = scorecard_tn_export   normalize_unitid(scorecard_tn_export['unitid'])
```

```python
#####
scorecard_tn_export['cip4'] = (
    scorecard_tn_export['cip4']
        .astype(str)
        .str.replace('.', '', regex=False)
        .str.strip()
        .str[:4]
        .str.zfill(4)
)
```

```python
#####
scorecard_tn_export['credlev_n'] = pd.to_numeric(
    scorecard_tn_export['credlev'], errors='coerce'
).astype('Int64')
```

```python
#####
scorecard_tn_export['credlev_str'] = scorecard_tn_export['credlev_n'].astype(str)
```

```python
scorecard_tn_export.head(1)
```

```python
#####
# Build composite key
scorecard_tn_export['composite_key'] = (
    scorecard_tn_export['unitid_n'] + "|" +
    scorecard_tn_export['cip4'] + "|" +
    scorecard_tn_export['credlev_str']
)
```

```python
scorecard_tn_export.to_csv(
    "scorecard_tn_export.csv",
    index=False,
    encoding="utf-8"
)
```
