```python
# Select columns to keep.
costs_2011_clean = costs_2011[[
    "UNITID"
    ,"TUITION2"
    ,"FEE2"
    ,"TUITION3"
    ,"FEE3"
    ,"CHG10"
    ,"CHG1AY1"
    ,"CHG1AY2"
    ,"CHG1AY3"
    ,"CHG2AY0"
    ,"
]].copy()
```

```python
costs_2011_clean["instate_cost"] = (
    costs_2011_clean["TUITION2"] + costs_2011_clean["FEE2"]
)
costs_2011_clean["outstate_cost"] = (
    costs_2011_clean["TUITION3"] + costs_2011_clean["FEE2"]
)
```

```python
costs_2011_clean.columns = costs_2011_clean.columns.str.lower()

for col in ["tuition2", "fee2", "tuition3", "fee3", "instate_cost", "outstate_cost"]:
    costs_2011_clean[col] = pd.to_numeric(
        costs_2011_clean[col],
        errors="coerce"
    )
```
